Задание №2

Написать генератор случайных матриц, который принимает
опциональные параметры m, n, min_limit, max_limit, где m и n указывают размер
матрицы, а min_lim и max_lim - минимальное и максимальное значение для
генерируемого числа .

```python
import random

def genMas(m=50, n=50, min_limit=250, max_limit=1010):
    mas = []
    for i in range(m):
        mas.append([0]*n)
    for i in range(m):
        for j in range(n):
            mas[i][j] = random.randint(min_limit, max_limit)
    return mas
```

Реализовать методы сортировки строк числовой матрицы в соответствии с заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со временем стандартной функции сортировки. Испытания проводить на сгенерированных матрицах.

```python
def Choice(b): #Сортировка выборкой
    a = b.copy()
    for i in range(len(a) - 1):
        m = i
        j = i + 1
        while j < len(a):
            if a[j] < a[m]:
                m = j
            j = j + 1
        a[i], a[m] = a[m], a[i]

def Insert(b): #Сортировка вставкой
    a = b.copy()
    for i in range(1, len(a)):
        k = a[i]
        j = i-1
        while j >= 0 and k < a[j]:
            a[j+1] = a[j]
            j -= 1
        a[j+1] = k

def Exchange(b): #Сортировка обменом
    a = b.copy()
    for i in range(len(a)):
        for j in range(len(a)-i-1):
            if a[j] > a[j+1]:
                a[j], a[j + 1] = a[j + 1], a[j]

def Shell(a): #Сортировка Шелла
    arr = a.copy()
    d = len(arr) // 2
    while d:
        for i, el in enumerate(arr):
            while i >= d and arr[i - d] > el:
                arr[i] = arr[i - d]
                i -= d
            arr[i] = el
        d = 1 if d == 2 else int(d * 5.0 / 11)

#Турнирная сортировка
def tournamentSort(array):
    arr = array.copy()
    for i in range(len(arr)):
        Tournament(arr[i])
    return arr


def Tournament(arr):
    tree = [None] * 2 * (len(arr) + len(arr) % 2)
    index = len(tree) - len(arr) - len(arr) % 2

    for i, v in enumerate(arr):
        tree[index + i] = (i, v)

    for j in range(len(arr)):
        n = len(arr)
        index = len(tree) - len(arr) - len(arr) % 2
        while index > -1:
            n = (n + 1) // 2
            for i in range(n):
                i = max(index + i * 2, 1)
                if tree[i] != None and tree[i + 1] != None:
                    if tree[i][1] < tree[i + 1][1]:
                        tree[i // 2] = tree[i]
                    else:
                        tree[i // 2] = tree[i + 1]
                else:
                    tree[i // 2] = tree[i] if tree[i] != None else tree[i + 1]
            index -= n

        index, x = tree[0]
        arr[j] = x
        tree[len(tree) - len(arr) - len(arr) % 2 + index] = None

#Пирамидальная сортировка
def Pyramid1(arr, n, i):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2

    if l < n and arr[i] < arr[l]:
        largest = l

    if r < n and arr[largest] < arr[r]:
        largest = r

    if largest != i:
        arr[i],arr[largest] = arr[largest],arr[i]

        Pyramid1(arr, n, largest)


def Pyramid2(array):
    arr=array.copy()
    for i in range(len(arr)):
        n = len(arr[i])

        for j in range(n, -1, -1):
            Pyramid1(arr[i], n, j)

        for j in range(n-1, 0, -1):
            arr[i][j], arr[i][0] = arr[i][0], arr[i][j]
            Pyramid1(arr[i], j, 0)
    return arr

#Быстрая сортировка
def Quick1(a):
    arr = a.copy()
    for i in range(len(arr)):
        Quick2(0, len(arr[i]) - 1, arr, i)
    return arr


def Quick2(fst, lst, array, row):
    first = int(fst)
    last = int(lst)
    middle = int((first + last) / 2)

    while (first < last):

        while (array[row][first] < array[row][middle]):
            first += 1
        while (array[row][last] > array[row][middle]):
            last -= 1
        if (first <= last):
            array[row][first], array[row][last] = array[row][last], array[row][first]
            first += 1
            last -= 1

    if (fst < last):
        Quick2(fst, last, array, row)
    if (first < lst):
        Quick2(first, lst, array, row)
```

```python
import timeit

matrix = genMas()

startTime = timeit.default_timer()
for i in range(len(matrix)):
    Choice(matrix[i])
print("Сортировка выборкой длится " + str(timeit.default_timer() - startTime) + " сек")

startTime = timeit.default_timer()
for i in range(len(matrix)):
    Insert(matrix[i])
print("Сортировка вставкой длится " + str(timeit.default_timer() - startTime) + " сек")

startTime = timeit.default_timer()
for i in range(len(matrix)):
    Exchange(matrix[i])
print("Сортировка обмена длится " + str(timeit.default_timer() - startTime) + " сек")

startTime = timeit.default_timer()
for i in range(len(matrix)):
    Shell(matrix[i])
print("Сортировка Шелла длится " + str(timeit.default_timer() - startTime) + " сек")

startTime = timeit.default_timer()
tournamentSort(matrix)
print("Турнирная сортировка длится " + str(timeit.default_timer() - startTime) + " сек")

startTime = timeit.default_timer()
Pyramid2(matrix)
print("Пирамидальная сортировка длитс " + str(timeit.default_timer() - startTime) + " сек")

startTime = timeit.default_timer()
Quick1(matrix)
print("Быстрая сортировка длится " + str(timeit.default_timer() - startTime) + " сек")

mas = matrix.copy()
startTime = timeit.default_timer()
mas.sort()
print("Встроенная сортировка длится " + str(timeit.default_timer() - startTime) + " сек")
```
